### Filtering using GPT4

```python
from openai import OpenAI

client = OpenAI(
    # defaults to os.environ.get("OPENAI_API_KEY")
    api_key=openai.api_key,
)

model = "gpt-3.5-turbo-0125"

def get_completion(prompt, model=model):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
    temperature=0.5
        )
    return response.choices[0].message.content.strip()
```

```python
from tqdm.auto import tqdm
import re

gpt_answers =[]
prompt = f"""
Provide at least 50 synonymous sentences for the following instruction: Convert the photo to Christmas style with holiday decorations.
Only print the output sentences without numbering.
"""

response = get_completion(prompt)
gpt_answers.append(response)
```

```python
paraphrasing = gpt_answers[0].split('\n- ')
paraphrasing
```
